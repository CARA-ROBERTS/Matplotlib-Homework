```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_mouse_data = pd.merge(mouse_drug_data, clinical_trial_data, on="Mouse ID")


# Display the data table for preview
combined_mouse_data.head()
```

```python
combined_mouse_data.dtypes
```

```python
combined_mouse_data['Timepoint']= combined_mouse_data.loc[:,'Timepoint'].astype(object)
combined_mouse_data['Timepoint']
```

## Metastatic Response to Treatment

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count =combined_mouse_data["Mouse ID"].count() 
mice_count
# Convert to DataFrame
mice_count_df = pd.DataFrame ({"Count of Mice" : mice_count})
# Preview DataFrame
mice_count_df
```

```python
# Generate the Plot (Accounting for percentages)
Time =[0,5,10,15,20,25,30,35,40,45]

Capomulin = Reformat_mice_count_df["Capomulin"]
Infubinol = Reformat_mice_count_df["Infubinol"] 
Ketapril = Reformat_mice_count_df["Ketapril"]
Placebo = Reformat_mice_count_df["Placebo"]

plt.plot(Time,Capomulin, marker = "o", color= "red", label="Capomulin")
plt.plot(Time, Infubinol,marker= "^",color="blue", label= "Infubinol")
plt.plot(Time,Ketapril,marker= "s",color="green", label= "Ketapril")
plt.plot(Time, Placebo,marker= "d",color="black", label= "Placebo")

plt.legend(loc="best")
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()

# Save the Figure
plt.savefig("Survival.png")
# Show the Figure
plt.show()
```
